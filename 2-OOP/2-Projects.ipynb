*Write a Python program that inputs a polynomial in standard algebraic
notation and outputs the first derivative of that polynomial.*

```python
class Polynomial:
    def __init__(self, degree, constants):
        self.constants = constants
        self.degree = degree
        assert len(constants) == degree + 1

    def first_derivative(self):
        deriv = ''
        for i in range(self.degree):
            if i == self.degree - 1: deriv += f'{constants[i] * (self.degree - i)}'
            elif self.degree - i == 2: deriv += f'{constants[i] * (self.degree - i)}x + '
            else: deriv += f'{constants[i] * (self.degree - i)}x^{self.degree - i - 1} + '
        return deriv

    def print_polynomial(self):
        poly = ''
        for i in range(self.degree+1):
            if i == self.degree: poly += str(constants[i])
            else: 
                if constants[i] == 1: poly += f'x^{self.degree - i} + '
                elif self.degree - i == 1: poly += f'{constants[i]}x + '
                else: poly += f'{constants[i]}x^{self.degree - i} + '
        return poly
```

```python
constants = [8,7,6,5,4,3,2,1]
degree = 7
p = Polynomial(degree, constants)
print(p.print_polynomial())
print(p.first_derivative())
```

üìå **Problem Index: P-2.34:**
*Write a Python program that inputs a document and then outputs a bar-chart plot of the frequencies of each alphabet character that appears in
that document.*

```python
class WordCount:
    def __init__(self, filename):
        self.filename = filename
        self.counter = dict()

    def count_words(self):
        with open(self.filename, 'r') as f:
            for row in f.readlines():
                for word in row.split(' '):
                    for char in word:
                        if char.lower() in self.counter: self.counter[char.lower()] += 1
                        else: self.counter[char.lower()] = 1
        return self.counter

filename = './words.txt'
wc = WordCount(filename)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
count = wc.count_words()

plt.figure(figsize=(15,7))
plt.bar(count.keys(), count.values())
plt.show()
```

*Write a set of Python classes that can simulate an Internet application in
which one party, Alice, is periodically creating a set of packets that she
wants to send to Bob. An Internet process is continually checking if Alice
has any packets to send, and if so, it delivers them to Bob‚Äôs computer, and
Bob is periodically checking if his computer has a packet from Alice, and,
if so, he reads and deletes it.*

```python
class User:
    def __init__(self, name):
        self.name = name
        self.anyMessages = False
        self.messages = []

    def receive_msg(self, sender, msg):
        self.messages.append(msg)
        self.anyMessages = True
        print(f'Message received from {sender}...')

    def read_msg(self):
        if len(self.messages) > 0:
            msg = self.messages[-1]
            self.messages.pop()
            print(f'Message> {msg}')
        else:
            self.anyMessages = False
            print('No messages left...')
```

*Develop an inheritance hierarchy based upon a Polygon class that has
abstract methods area( ) and perimeter( ). Implement classes Triangle,
Quadrilateral, Pentagon, Hexagon, and Octagon that extend this base
class, with the obvious meanings for the area( ) and perimeter( ) methods.
Also implement classes, IsoscelesTriangle, EquilateralTriangle, Rectangle, and Square, that have the appropriate inheritance relationships. Finally, write a simple program that allows users to create polygons of the
various types and input their geometric dimensions, and the program then
outputs their area and perimeter. For extra effort, allow users to input
polygons by specifying their vertex coordinates and be able to test if two
such polygons are similar.*

```python
from abc import abstractmethod, ABCMeta

class Polygon(metaclass=ABCMeta):
    def __init__(self, sides=[1,1,1], n_sides=3):
        self._sides = sides
        self._n_sides = n_sides
    
    @abstractmethod
    def area():
        pass

    @abstractmethod
    def perimeter():
        pass
```

```python
import math
```

```python
class Triangle(Polygon):
    def __init__(self, sides):
        super().__init__(sides, 3)
        self._perimeter = self.perimeter()
        self._area = self.area()
    def perimeter(self):
        return sum(self._sides)
    def area(self):
        s = self._perimeter/2
        product = s
        for i in self._sides:
            product *= (s-i)
        return product**0.5

class IsoscelesTriangle(Triangle):
    def __init__(self, b, h):
        self._b = b
        self._h = h
        self._perimeter = self.perimeter()
        self._area = self.area()
    def perimeter(self):
        iso_sides = 2 * (self._h ** 2 + (self._b**2) / 4)
        return iso_sides + self._b
    def area(self):
        return 0.5 * self._b * self._h

class EquilateralTriangle(Triangle):
    def __init__(self, a):
        self._a = a
        self._perimeter = self.perimeter()
        self._area = self.area()
    def perimeter(self):
        return 3*a
    def area(self):
        return ((3)**0.5 * a * a) / 4
```

```python
class Quadrilateral(Polygon):
    def _init__(self, sides):
        super().__init__(sides, 4)
        self._sides = sides
        self._perimeter = self.perimeter()
    def perimeter(self):
        return sum(self._sides)

class Square(Quadrilateral):
    def __init__(self, a):
        self._side = a
        self._perimeter = self.perimeter()
        self._area = self.area()
    def perimeter(self):
        return 4*self._side
    def area(self):
        return self._side**2

class Rectangle(Quadrilateral):
    def __init__(self, a, b):
        self._a = a
        self._b = b
        self._perimeter = self.perimeter()
        self._area = self.area()
    def perimeter(self):
        return 2*(self._a + self._b)
    def area(self):
        return self._a * self._b
```
